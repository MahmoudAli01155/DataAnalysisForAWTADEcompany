```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\pc\Downloads\output4 (1).xlsx")
print(df)
```

```python
print(df.info())  
```

```python
df.rename(columns={"OrderSatus": "OrderStatus", "GPSlongitude": "GPSLongitude"}, inplace=True)
print(df)
```

```python
df.drop_duplicates(inplace=True)
print(df)
```

```python
df.loc[(df["GPSLatitude"] == 0) & (df["GPSLongitude"] == 0), ["GPSLatitude", "GPSLongitude"]] = None
print(df)
```

```python
date_cols = ['OrderDate', 'DesiredDeliveryDate', 'CreatedDate', 'UpdatedDate']
df[date_cols] = df[date_cols].apply(pd.to_datetime, errors='coerce')
print(df)
```

```python
df['GrossTotal'] = df['GrossTotal'].astype(float).round(2)
print(df)
```

```python
df.to_excel("C:\\Users\\pc\\Downloads\\cleaned_output.xlsx", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='CreatedBy', order=df['CreatedBy'].value_counts().index, palette="viridis")
plt.title('Number of Orders per Creator')
plt.xlabel('Created By')
plt.ylabel('Number of Orders')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['GrossTotal'], bins=20, kde=True, color="skyblue")
plt.title('Gross Total Distribution')
plt.xlabel('Gross Total')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
